```python
L, R, d = map(int, input().split())

cnt = 0
for n in range(L, R+1):
    if n % d == 0:
        cnt += 1

print(cnt)
```

```python
N = int(input())
A = list(map(int, input().split()))

cnt = 0
for i in range(N):
    if i % 2 == 0:
        if A[i] % 2 != 0:
            cnt += 1
    else:
        continue    
print(cnt)
```

```python
#高速だけど10**4だけ通せない
import itertools
N = int(input())

ans = [0]*(10**5)
for i1 in range(1, 40):
    i2 = i1+1
    i3 = i1+2
    if i3 > N:
        break
    L = list(itertools.combinations_with_replacement((i1, i2, i3), 3))
    for i in L:
        i = list(i)
        x = i[0]
        y = i[1]
        z = i[2]
        n = (x**2)+(y**2)+(z**2)+(x*y)+(y*z)+(z*x)
        if len(list(set(i))) == 1:
            ans[n] += 1
        if len(list(set(i))) == 2:
            ans[n] += 3
        if len(list(set(i))) == 3:
            ans[n] += 6
        
for k in range(1, N+1):
    print(ans[k])
```

```python
#解答例
n = int(input())
ans = [0 for _ in range(10050)]
for i in range(1,105):
    for j in range(1,105):
        for k in range(1,105):
            v = i*i+j*j+k*k+i*j+j*k+k*i
            if v<10050:
                ans[v]+=1
for i in range(n):
    print(ans[i+1])
```
